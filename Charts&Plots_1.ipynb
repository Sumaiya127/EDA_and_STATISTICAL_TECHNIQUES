```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

#LINE PLOT

plt.figure(figsize=(6,4))
plt.plot(x, y, label='y = sin(x)')
plt.title('Line Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()
plt.show()
```

```python
##HISTOGRAM

data = np.random.randn(1000)
plt.figure(figsize=(6,4))
plt.hist(data, bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
##BAR CHART

labels = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 30]
plt.figure(figsize=(6,4))
plt.bar(labels, values)
plt.title('Bar Chart')
plt.xlabel('Labels')
plt.ylabel('Values')
plt.show()
```

```python
##SCATTER PLOT

x_vals = np.random.rand(100)
y_vals = np.random.rand(100)
plt.figure(figsize=(6,4))
plt.scatter(x_vals, y_vals, color='blue')
plt.title('Scatter Chart')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

```python
##PIE CHART

sizes = [215, 130, 245, 210]
labels = ['A', 'B', 'C', 'D']
explode = (0.1, 0, 0, 0)  # explode 1st slice for emphasis
plt.figure(figsize=(6,4))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Pie Chart')
plt.show()
```
